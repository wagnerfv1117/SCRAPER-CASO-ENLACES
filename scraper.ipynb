```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL de la página web que queremos escrapear
url = "https://www.larepublica.co/"

# Hacer la solicitud HTTP a la página web
respuesta = requests.get(url)
print("La respuesta del servidor es:", respuesta)
```

```python
with open(path + 'extraccion.txt', 'wb') as output:
    output.write(respuesta.content)
```

```python
# Crear objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(respuesta.content, 'html.parser')

# Buscar todos los elementos <a> y extraer el texto y los enlaces
for enlace in soup.find_all('a'):
    texto = enlace.text.strip() # Extraer el texto del enlace
    url = enlace.get('href') # Extraer la URL del enlace
    print(texto, url) # Imprimir el texto y la URL
```

```python
# Lista para almacenar los datos extraídos
datos = []

# Buscar todos los elementos <a> y extraer el texto y los enlaces
for enlace in soup.find_all('a'):
    texto = enlace.text.strip() # Extraer el texto del enlace
    url = enlace.get('href') # Extraer la URL del enlace
    
    # Añadir los datos a la lista como un diccionario
    datos.append({'Titulo': texto, 'Enlace URL': url})

# Crear el dataframe a partir de la lista de diccionarios
df = pd.DataFrame(datos)
```
